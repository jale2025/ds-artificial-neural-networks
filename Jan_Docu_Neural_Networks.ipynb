### 1. Die Architektur definieren (Das Skelett)

Der gängigste Weg ist das Sequential-Modell. Hier definierst du die Schichten nacheinander.

```python
import tensorflow as tf
from tensorflow.keras import layers

# Ein lineares Stapel-Modell erstellen
model = tf.keras.Sequential([
    # Eingabeschicht: Erwartet z.B. 8 Merkmale (Features)
    layers.Input(shape=(8,), name= 'input_layer'), 
    
    # Erste versteckte Schicht mit 16 Neuronen und ReLU-Aktivierung
    layers.Dense(units=16, activation='relu', name= 'hidden_layer_1', kernel_initializer="uniform"),
    
    # Zweite versteckte Schicht
    layers.Dense(units=8, activation='relu', name= 'hidden_layer_2', kernel_initializer="uniform"),
    
    # Ausgabeschicht: 1 Neuron für eine Vorhersage (z.B. Preis oder Ja/Nein)
    layers.Dense(units=1, activation='sigmoid') 
])
```

```python
# Alternativ kann man auch die add Methode verwenden
model = keras.models.Sequential() # initialize the model

model.add(keras.layers.Dense(input_dim = 2, # only needed for the first layer
                            units = 2,      # number of neurons
                            activation = keras.activations.sigmoid,
                            kernel_initializer="uniform",
                            name= 'hidden_layer'))

model.add(keras.layers.Dense(units = 1,
                            activation = keras.activations.sigmoid,
                            name= 'output_layer'))
```

1. Grundlegende Parameter

    units:
    Gibt die Anzahl der Neuronen in dieser Schicht an. Dies bestimmt die Dimensionalität des Outputs.

    activation:
    Die Aktivierungsfunktion, die auf den Output angewendet wird (z. B. 'relu', 'sigmoid', 'softmax'). Ohne Angabe bleibt der Output linear (f(x)=x).

2. Initialisierung (Wie fängt das Modell an?)

    Diese Parameter bestimmen, mit welchen Startwerten die Gewichte und Biases belegt werden, bevor das Training beginnt.

    use_bias (Boolean):
    Legt fest, ob die Schicht einen Bias-Vektor (einen konstanten Versatzwert) verwendet. Standard ist True.

    kernel_initializer:
    Bestimmt die Startwerte für die Gewichte (Kernel). Beliebte Optionen sind 'glorot_uniform' oder 'he_normal'.

    bias_initializer:
    Bestimmt die Startwerte für den Bias. Standardmäßig meist auf Null gesetzt ('zeros')

3. Regularisierung (Vermeidung von Overfitting)

    Diese Parameter helfen dabei, das Modell zu "bestrafen", wenn die Gewichte zu groß oder zu komplex werden, um die Generalisierung zu verbessern.

    kernel_regularizer:
    Wendet eine Strafe auf die Gewichte an (z. B. L1 oder L2 Regularisierung).

    bias_regularizer:
    Wendet eine Strafe auf den Bias-Vektor an.

    activity_regularizer:
    Wendet eine Strafe direkt auf den Output der Schicht (die Aktivierung) an.

4. Constraints (Einschränkungen)

    Hier kannst du erzwingen, dass die Gewichte während des gesamten Trainings bestimmte Bedingungen erfüllen.

    kernel_constraint:
    Ermöglicht es, die Gewichte einzuschränken (z. B. NonNeg(), um nur positive Gewichte zuzulassen, oder MaxNorm()).

    bias_constraint:
    Dasselbe Prinzip wie oben, angewendet auf den Bias.

optimizer: 
Der Optimizer ist der Algorithmus, der die Gewichte des neuronalen Netzes anpasst, um den Fehler zu minimieren. Er bestimmt wie das Modell lernt.
Er berechnet, wie die internen Parameter (Gewichte) basierend auf dem Gradienten der Loss-Funktion verändert werden müssen.
    Moegliche Werte: Adam/SGD/LossScaleOptimizer

loss: 
Die Loss-Funktion ist das Maß für den Fehler. Sie sagt dem Modell, wie weit seine Vorhersage vom tatsächlichen Zielwert entfernt ist. Das Ziel des Trainings ist es, diesen Wert so nah wie möglich an Null zu bringen. Je kleiner der Loss, desto besser die Vorhersage. 
Moegliche Werte: 
- Binaere Klassifikation: BinaryCrossentropy
- Multi-Klassifikation: categorical_crossentropy oder sparse_categorical_crossentropy
- Regression: MeanAbsoluteError/MeanSquaredError

metrics: 
Metriken dienen der Überwachung des Trainingserfolgs durch den Menschen. Im Gegensatz zum Loss werden Metriken nicht direkt für das Training (die mathematische Optimierung) verwendet. Sie helfen dir zu beurteilen, wie gut dein Modell in der "echten Welt" performt. Metriken sind oft leichter zu interpretieren als der Loss-Wert.
    Moegliche Werte: R2Score/AUC/Accuracy/Recall/F1Score/FalseNegatives/FalsePositives/MeanSquaredError/RootMeanSquaredError

    Anzahl der Daten(zeilen), die pro Durchgang als Trainingsdaten verwendet werden. Angenomen wir haben 1000 Zeilen und wir haben die batch_size = 100, dann besteht eine epoch aus 10 Durchgängen. epoch:
    Ein vollständiger Durchlauf des gesamten Trainingsdatensatzes durch das Netzwerk. Während einer Epoche wird jedes einzelne Beispiel aus deinem Trainingsset genau einmal vom Modell gesehen. 
    Dieser Prozess umfasst normalerweise:
    - Forward Pass: Die Daten fließen durch das Netz, und eine Vorhersage wird getroffen.

    Warum reicht eine Epoche nicht aus?:
    In der Regel trainiert man ein Modell über viele Epochen (z. B. 10, 50 oder 500). Da die Gewichte bei einem Durchlauf meist nur minimal angepasst werden, muss das Modell die Daten mehrfach sehen, um komplexe Muster zu "erlernen".

    Zu wenige Epochen: Das Modell lernt nicht genug (Underfitting).
    Zu viele Epochen: Das Modell merkt sich die Daten auswendig und verliert die Fähigkeit zur Generalisierung (Overfitting)

### 6. convolutional layer

#### Convolution
- Use Kernel (small matrix of weights) to slide over the 2D image
- Perform element wise multiplication with the overlapping part
- Sum up the results into a single output pixel

<img src="https://miro.medium.com/max/1400/1*Fw-ehcNBR9byHtho-Rxbtw.gif" alt="Drawing" style="width: 400px;"/>

#### Padding (Auffüllen)

Beim Filtern eines Bildes (Convolution) passiert normalerweise Folgendes: Die Ränder werden seltener gescannt als die Mitte, und das Bild wird mit jeder Schicht kleiner. Padding löst dieses Problem.

Was es macht: Es fügt einen künstlichen Rand aus Pixeln (meistens mit dem Wert 0, daher „Zero-Padding“) um das Bild herum hinzu.
Warum man es nutzt:
Größenerhalt: 
- Man verhindert, dass das Bild durch die Convolution schrumpft.
- Randinformationen: Die Pixel am Rand werden genauso gewichtet wie die in der Mitte.

Wichtige Modi in Keras:
- valid: Kein Padding (Bild wird kleiner).
- same: Padding wird so berechnet, dass der Output dieselbe Größe wie der Input hat.

#### Stride (Schrittweite)

Stride definiert, wie weit sich der Filter (Kernel) bei jedem Schritt über das Bild bewegt.

- Was es macht: Ein Stride von 1 bedeutet, der Filter rückt einen Pixel nach dem anderen vor. Ein Stride von 2 bedeutet, er springt immer zwei Pixel weiter.

- Effekt: Je größer der Stride, desto kleiner wird das Ergebnisbild. Es ist eine Methode, um die Datenmenge zu reduzieren, während der Filter über das Bild „hüpft“.

- Standard: In den meisten Fällen ist der Stride auf (1, 1) gesetzt.

#### Pooling (Zusammenfassung)

Pooling-Schichten dienen dazu, die räumliche Größe der Daten zu verkleinern, um Rechenleistung zu sparen und das Modell robuster gegen kleine Verschiebungen im Bild zu machen.

Was es macht: Es fasst einen Bereich (z. B. 2×2 Pixel) zu einem einzigen Wert zusammen.

Gängige Arten:
- Max Pooling: Nur der höchste Wert aus dem Bereich wird weitergegeben (wichtigstes Merkmal).
- Average Pooling: Der Durchschnittswert des Bereichs wird berechnet.

Warum man es nutzt: Es reduziert die Anzahl der Parameter (verhindert Overfitting) und macht das Modell „ortsunabhängiger“ (egal ob das Katzenohr 2 Pixel weiter links oder rechts ist).

| Begriff | Hauptaufgabe | Auswirkung auf die Bildgröße |
| :--- | :--- | :--- |
| **Padding** | Ränder schützen / Größe erhalten | Vergrößert den Input (vor der Convolution) |
| **Stride** | Geschwindigkeit des Scannens | Verkleinert den Output drastisch bei Werten > 1 |
| **Pooling** | Datenreduktion & Abstraktion | Verkleinert den Output (meistens um die Hälfte) |

### 7. Overfit and Underfit

You have already learned about overfitting and underfitting in the context of machine learning algorithms. Unfortunately, these topics are also important when it comes to training an artificial neural network.

You may have observed in previous notebooks that the accuracy(/mae) of your model on the validation set reaches a peak(/minimum) after training for a number of epochs and then stagnates or, worse, starts to decrease(/increase). This is a clear sign that the model is overfitting the training data. 

Artificial neural networks tend to overfit the data. It is possible to achieve extremely high accuracy (or low mae) on your training data if you just train your model long enough. However, this is not desirable because these models do not generalize well and will perform poorly on a test set or new data.

Underfitting is another problem. If your model is not powerful enough, it is over-regularized, or it has not been trained long enough, it may perform poorly on the training and test set because it has not been able to learn the relevant features from the data.

The best way to prevent your model from overfitting is using a dataset for training that covers the full range of inputs that the model is expected to handle. Additional data may only be useful if it covers new and interesting cases.

However, sometimes you will need to use different regularization techniques to counteract the overfitting of the data. In this notebook we will explore some common techniques that can help you with this task.

Strategies against overfitting:
- Get more training data.
- Reduce the capacity of the network.
- Add weight regularization -> L1 (Lasso: reduces some features to 0) or L2 (Ridge: reduce all parameter a bit) regularization
- Add dropout: Set randomly some features to zero in each iteration of a epoch

Two important approaches not covered in the bootcamp:

* data-augmentation
* batch normalization

Remember that each method can help on its own, but often combining them can be even more effective.


## Strategies to prevent overfitting

### Regularization 

You may be familiar with Occam's Razor principle: given two explanations for something, the explanation most likely to be correct is the "simplest" one, the one that makes the least amount of assumptions. This also applies to the models learned by neural networks: given some training data and a network architecture, there are multiple sets of weights values (multiple models) that could explain the data, and simpler models are less likely to overfit than complex ones.

In the next section we will try out one form of **regularization**. Hopefully, you all are familiar with L1 and L2 regularization as we talked about it several times during the bootcamp. Nevertheless, let's refresh the general concepts before we move on: 

We can apply regularization techniques in order generate a simpler model to prevent overfitting. A "simple model" in this context is a model where the distribution of parameter values has less entropy (or a model with fewer parameters altogether). Thus a common way to mitigate overfitting is to put constraints on the complexity of a network by forcing its weights only to take small values, which makes the distribution of weight values more "regular". It is done by adding to the loss function of the network a cost associated with having large weights. This cost comes in two flavors:

* [L1 regularization](https://developers.google.com/machine-learning/glossary/#L1_regularization), where the cost added is proportional to the absolute value of the weights coefficients (i.e. to what is called the "L1 norm" of the weights).

* [L2 regularization](https://developers.google.com/machine-learning/glossary/#L2_regularization), where the cost added is proportional to the square of the value of the weights coefficients (i.e. to what is called the squared "L2 norm" of the weights). L2 regularization is also called weight decay in the context of neural networks. Don't let the different name confuse you: weight decay is mathematically the exact same as L2 regularization.

L1 regularization pushes weights towards exactly zero encouraging a sparse model. L2 regularization will penalize the weights parameters without making them sparse since the penalty goes to zero for small weights - one reason why L2 is more common.

### Dropout

Dropout is one of the most effective and most commonly used regularization techniques for neural networks, developed by Hinton and his students at the University of Toronto.

The intuitive explanation for dropout is that because individual nodes in the network cannot rely on the output of the others, each node must output features that are useful on their own.

Dropout, applied to a layer, consists of randomly "dropping out" (i.e. set to zero) a number of output features of the layer during training. Let's say a given layer would normally have returned a vector [0.2, 0.5, 1.3, 0.8, 1.1] for a given input sample during training; after applying dropout, this vector will have a few zero entries distributed at random, e.g. [0, 0.5, 1.3, 0, 1.1].

The "dropout rate" is the fraction of the features that are being zeroed-out; it is usually set between 0.2 and 0.5. At test time, no units are dropped out, and instead the layer's output values are scaled down by a factor equal to the dropout rate, so as to balance for the fact that more units are active than at training time.

In `tf.keras` you can introduce dropout in a network via the Dropout layer, which gets applied to the output of layer right before.

Let's add a Dropout layer after each dense layer in our network to see how they influence our models performance.

Dropout usually slows convergence a bit because the network sees a noisier view of itself each batch. That's expected—look for validation curves that are smoother and peak later, indicating better generalization. You can sweep rates like 0.2, 0.3, 0.5 to see the trade-off between regularization strength and training speed.

```python
# Large model with additional dropout layer
def get_compiled_dropout_model():
    dropout_model = tf.keras.Sequential(
        [
            tf.keras.layers.Dense(
                512, kernel_initializer="uniform", activation="relu", input_dim=12
            ),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(512, kernel_initializer="uniform", activation="relu"),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(512, kernel_initializer="uniform", activation="relu"),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(512, kernel_initializer="uniform", activation="relu"),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(1, kernel_initializer="uniform"),
        ]
    )

    dropout_model.compile(optimizer=optimizer, loss="mae", metrics=["mse"])
    return dropout_model
```

### Combine L2 and Dropout

It's also possible and common to combine different regularization techniques in one model. 

Combining both lets L2 keep weights small while dropout combats co-adaptation between neurons. Start with the same dropout rate you found reasonable above and a slightly smaller L2 coefficient to avoid over-regularizing.

```python
# Large model with additional dropout layer and L2 regularization
def get_compiled_l2_dropout_model():
    l2_dropout_model = tf.keras.Sequential(
        [
            tf.keras.layers.Dense(
                512,
                kernel_initializer="uniform",
                activation="relu",
                kernel_regularizer=regularizers.l2(0.01),
                input_dim=12,
            ),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(
                512,
                kernel_initializer="uniform",
                activation="relu",
                kernel_regularizer=regularizers.l2(0.01),
            ),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(
                512,
                kernel_initializer="uniform",
                activation="relu",
                kernel_regularizer=regularizers.l2(0.01),
            ),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(
                512,
                kernel_initializer="uniform",
                activation="relu",
                kernel_regularizer=regularizers.l2(0.01),
            ),
            tf.keras.layers.Dropout(0.25),
            tf.keras.layers.Dense(1, kernel_initializer="uniform"),
        ]
    )

    l2_dropout_model.compile(optimizer=optimizer, loss="mae", metrics=["mse"])
    return l2_dropout_model
```

### ANN Hyperparameter
Here a short summary about some hyperparameters in ANN we touch this afternoon:
- [Important activation functions](https://medium.com/hyunjulie/activation-functions-a-short-summary-8450c1b1d426) and their usage:  
  - step function --> simple but has zero gradient everywhere  
  - linear function --> could be used for scaling but does not provide any non-linearity  
  - sigmoid --> mostly used in classification problems as output layer; suffers from vanishing gradient problem, not zero-centered  
  - tanh --> zero-centered and steeper gradient than sigmoid; mostly used in hidden layers but again vanishing gradient problem  
  - ReLU, leaky ReLU --> standard activation function; ReLU has no gradient for negative inputs  
  - softmax --> used for multiclass output as it sums up to 1 so that each value gives the probability of each class  
  
- [Some optimizers](https://shekhar-banerjee96.medium.com/mastering-gradient-descent-a-deep-dive-into-rmsprop-and-adam-optimizers-c57599b23be3):   
  - mini-bach SGD (stochastic gradient descent) --> uses only several observations to calculate the gradient and thus quite economical  
  - RMSprop --> adjusts the learning rate for each parameter individually: Steep gradients are slowed down, gentle gradients are given a boost  
  - Adam (Adaptive Moment Estimation) --> also uses momentum (portions of previous gradients) AND rmsprop; first choice  

- [Regularization](https://medium.com/ai%C2%B3-theory-practice-business/dropout-early-stopping-188a23beb2f9)  
  - L1/L2 --> well known: ass a penalty to the loss and work on the NN weights  
  - Dropout --> randomly deactivates a fraction of neurons during training, forcing the network to learn more robust features and preventing co-adaptation of neurons, thus reducing overfitting  
  - early stopping --> data is split into a training set and a validation set; Training is halted when the validation error stops improving  

- [Avoid vanishing gradients](https://www.youtube.com/watch?v=ncTHBi8a9uA) (many layers means multiplying many small numbers leading to a vanishing product)  
  - select ReLu instead of sigmoid --> gradient of 1 for positiv inputs.
  - batch normalization --> acts in between the 2 functions of a neuron (summation and activation function) by a kind of standard scaling  
  - weight initialization --> don't initially select weights randomly but with a certain distribution
  - create connections which bridge some layers --> architectural methods
